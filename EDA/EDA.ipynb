```python
'''
Desarrollo una función que me permita contruir un dataframe
con los valores no nulos, nulos, vacíos y distintos de una tabla
'''
def tablas(data):
  lista = []
  for columna in data.columns:
    No_nulos = data[columna].count()
    Nulos = data[columna].isnull().sum()
    Distintos = len(data[columna].unique())
    Vacíos = len(data[data[columna]==''])

    tupla = (columna, No_nulos, Nulos, Distintos, Vacíos)
    lista.append(tupla)
  df_ = pd.DataFrame(lista)
  df_.columns = ['Columna', 'No nulos', 'Nulos', 'Distintos', 'Vacíos']
  return df_

# Aplico la función

print(tablas(df))
```

```python
# Busqueda de valores únicos de cada variable
for i in df.columns:
  print(i,':', df[i].unique())
  print('--------------------')
```

```python
# Correción de datos en la variable 'age'
df['age'] = df['age'].replace('Dec-18','0-18')
# Reemplazo de valores por unos más cortos y acordes
df['occupation'] = df['occupation'].replace('Medical Professional aiding efforts against COVID-19', 'Medical Professional')
df['prefer'] = df['prefer'].replace({'Complete Physical Attendance':'Physical', 'Work/study from home':'Remote'})
df
```

```python
# Descartamos variables irrelevantas, la mayoría de los datos son nulos
df = df.drop(['line_of_work', 'Unnamed: 19', 'time_bp.1', 'travel+work'], axis=1)
```

```python
# Recuento de cada valor de cada variable
for i in df.columns:
  print('Columna:', i)
  print(df[i].value_counts())
  print('-----------------')
```

## Visualizaciones

```python
# Calcular la matriz de correlación
corr = df.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(12, 8))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('Correlation Matrix')
plt.show()
```

# Observaciones

Distribución por Género: La mayoría de los encuestados son hombres, representando el 55.2% del total, mientras que las mujeres representan el 44.4%. Existe un porcentaje muy pequeño (0.4%) que se identifica como 'Otros'.

Preferencia por el Modo de Asistencia: Una gran proporción de los encuestados (71.1%) prefieren la asistencia física completa, mientras que el 28.9% prefieren trabajar desde la casa.

Distribución de Ocupaciones: La ocupación más común entre los encuestados es 'Profesional Trabajando', seguida por 'Estudiante Universitario'.

Distribución del Tiempo Antes de la Pandemia: El tiempo más común antes de la pandemia es alrededor de 7 horas, con un número significativo de encuestados también pasando alrededor de 9 horas.

Distribución del Tiempo Durante la Pandemia: El tiempo más común durante la pandemia es de 9 horas, con un número significativo de encuestados también pasando alrededor de 7 horas.

Cambios en la distribución del tiempo: La distribución del tiempo de trabajo varía según las ocupaciones. Por ejemplo, para 'Homemaker' el tiempo de trabajo practicamente se duplicó durante la pandemia. En otros casos varió muy poco.

Distribución del Tiempo de Viaje: La mayoría de los encuestados pasan media hora viajando.

Distribución de la Puntuación del Entorno del Hogar: La puntuación más común para el entorno del hogar es 3, lo que sugiere un nivel moderado de comodidad.

Distribución del Incremento de Productividad: La mediana del puntaje de aumento de productividad es 0, con algunos valores atípicos en ambos lados. Esto sugiere que la mayoría de los encuestados no experimentaron un cambio significativo en su productividad.

Distribución del Equilibrio del Sueño: La mediana del puntaje de equilibrio del sueño es 0, con algunos valores atípicos en ambos lados. Esto sugiere que la mayoría de los encuestados no experimentaron un cambio significativo en su equilibrio del sueño.

Distribución de la Adquisición de Nuevas Habilidades: La mediana del puntaje de adquisición de nuevas habilidades es 0.5, con algunos valores atípicos en ambos lados. Esto sugiere que algunos encuestados pudieron adquirir nuevas habilidades.

Distribución de la Conexión Familiar: La mediana del puntaje de conexión familiar es 0.5, con algunos valores atípicos en el lado negativo. Esto sugiere que algunos encuestados pudieron aumentar sus conexiones familiares.

Distribución de la Relajación: La mediana del puntaje de relajación es 0, con algunos valores atípicos en ambos lados. Esto sugiere que la mayoría de los encuestados no experimentaron un cambio significativo en su nivel de relajación.

Distribución del Tiempo Personal: La mediana del puntaje de tiempo personal es 0, con algunos valores atípicos en ambos lados. Esto sugiere que la mayoría de los encuestados no experimentaron un cambio significativo en su tiempo personal.

Matriz de Correlación: Hay algunas correlaciones moderadas en los datos. Por ejemplo, travel_time está correlacionado positivamente con time_bp y time_dp, lo cual tiene sentido ya que más tiempo de viaje probablemente aumentaría el tiempo total pasado antes y durante la pandemia. Del mismo modo, home_env está correlacionado positivamente con easeof_online, lo que indica que un mejor entorno doméstico podría facilitar el trabajo en línea. Otra correlación que tiene sentido es relaxed con variables como sleep_bal, fam_connect y self_time. Sin embargo, no hay correlaciones muy fuertes (por encima de 0.7 o por debajo de -0.7), lo que sugiere que la mayoría de estas variables son relativamente independientes entre sí.

Ocupación por Género: La ocupación de 'Profesional Trabajando' tiene la mayor representación tanto de hombres como de mujeres. Le sigue 'Estudiante Universitario' con una predominancia de estudiantes masculinos. Para otras categorías, la distribución de género tiende hacia un género.

Preferencia por el Modo de Asistencia por Género: Tanto hombres como mujeres predominantemente prefieren la 'Asistencia Física Completa'. Sin embargo, una minoría considerable en ambos géneros prefiere una 'Mezcla de Asistencia en Línea y Física'. Además, la proporción mayoritaria de quienes prefieren 'Asistencia Física Completa' son hombres, mientras que la proporción mayoritaria que prefieren trabajo remoto son mujeres.

Puntuación del Entorno del Hogar por Género: La distribución de las puntuaciones del entorno del hogar es bastante similar tanto para hombres como para mujeres, con la mayoría de las personas puntuando su entorno hogareño alrededor del punto medio de la escala.
